```python
import pandas as pd
import numpy as np
import os
import glob
import pytz
from datetime import datetime
from google.colab import drive
drive.mount('/content/drive')
```

```python
path1="/content/drive/MyDrive/Risk Project/result/CS"
```

```python
CS = glob.glob(os.path.join(path1, "*.csv")) #make list of paths
CS_count = len(CS)
CS_list = []

for file in CS:
  temp_df = pd.read_csv(file)
  temp_df = temp_df.drop(columns =['RDSkew',	'down_vol_ratio',	'late_trans_ratio',	'vol_price_corr',	'avg_trans_outflow_ratio',	'large_order_drives_gain'])
  temp_df = temp_df.loc[temp_df['date']<'2022-01-01']
  CS_list.append(temp_df)

for dataset in CS_list:
  consumerStaple =pd.concat(CS_list)
  consumerStaple= consumerStaple[ ~( consumerStaple.isin( [np.inf, -np.inf]).any(axis =1) )] 
display(consumerStaple)

```

```python
consumerStaple =consumerStaple.groupby(by =['date']).mean()
consumerStaple = consumerStaple.loc[(consumerStaple.index == '2021-01-04')|(consumerStaple.index == '2021-04-01')
                  |(consumerStaple.index == '2021-07-01')|(consumerStaple.index == '2021-10-01')]
consumerStaple['industry'] = 'ConsumerStaple'
consumerStaple
```
