```python
# convert decimal integer to boolean array
def decimal_to_bool(decimal_number):
    boolean_array = []
    while decimal_number > 0:
        boolean_array.append(decimal_number % 2)
        decimal_number = decimal_number // 2
    boolean_array.reverse()
    return boolean_array
    
```
